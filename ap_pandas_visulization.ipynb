````python
import pandas as pd
import numpy as np

df = pd.read_csv('clean_amazon_data.csv')
df.head()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df.info()
````

Bar chart is used for categorical data like days Monday/Tuesday/Wednesday etc.

Line chart is useful when there is a comparison or you can see based on the year how the sales are changing then the line chart for each year makes sense

Histogram is meant for analyzing the single column

When you are dealing with two numeric column most of the time scatter plot is right choice.

Line chart makes sense when you have one column with constant increase value on x asis

Ask two questions to identify the right type of graph -
(1) Data type of the column
(2) How many columns

Boxplot - used when you have bivariate comparison -1 numeric versus 1 categorical value

````python
df['rating'].plot(kind='hist')
````

````python
mean_rating = df['rating'].mean()
plt.figure(figsize=(6, 4))
plt.hist(df['rating'].dropna(), bins=20, color='skyblue')
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rating',bins=20 , kde=False, color='skyblue', hue='contains_durable') # h.w experiment with binwidth.

# Styling Tip 💡
# Add a vertical line for the mean rating
mean_rating = df['rating'].mean()
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")

plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
````

````python
# category column
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='category', palette='coolwarm')

# Styling Tip 💡
# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

plt.title("Count of Products by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
````

````python
#To use the matplotlib to show the bar plot we have to first do the value counts for each category because matpotlib does not create the frequncy on y axis automagically

categories = df['category'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(categories.index, categories.values, color='skyblue')
plt.xticks(rotation=45)
plt.title("Count of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
````

````python

plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='rating',bins=20 , kde=True, color='skyblue') # h.w experiment with binwidth.

# Styling Tip 💡
# Add a vertical line for the mean rating
mean_rating = df['rating'].mean()
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")

plt.title("Distribution of Ratings", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
````

````python
# kde plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='discounted_price', shade=True, color='purple')

# Styling Tip 💡
# Add a vertical grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.title("KDE Plot of Discounted Prices", fontsize=16)
plt.xlabel("Discounted Price", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
````

````python
#actual price versus rating
sns.scatterplot(data=df, x='actual_price', y='discounted_price')
````

````python
sns.scatterplot(data=df, x='actual_price', y='discounted_price', hue = 'contains_durable')
````

````python
#average ratings over months
monthly_rating = df.groupby('month_year')['rating_filled'].mean().reset_index()

monthly_rating.info()

monthly_rating['month_year'] = pd.to_datetime(monthly_rating['month_year'], format='%B %Y')

monthly_rating.sort_values('month_year',inplace=True)
````

````python
sns.lineplot(data=monthly_rating, x='month_year', y='rating_filled')
````

Median - Median value is the middle value in the ascending or descnding sorted range

````python
#Bivariate analysis : 1 numeric versus 1 categorical value
#how are the ratings across various categories

sns.boxplot(data=df, x='category' , y='rating')
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='rating', palette='Set3') # Styling Tip 💡 palette

plt.xticks(rotation=45)
plt.title("Rating Distribution by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='actual_price', palette='Set3') # Styling Tip 💡 palette

plt.xticks(rotation=45)
plt.title("Rating Distribution by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
````

````python
#avg_rating, categories, durable vs non-durable #1 numeric and 2 categorical

df_cat = df.groupby(['category','contains_durable'])['rating'].mean().reset_index()

df_cat
````

````python
sns.barplot(data=df_cat, x='category', y='rating', hue='contains_durable' ) #grouped - bar chart
plt.xticks(rotation=45)
plt.show()
````

````python
#numeric versus numeric over multiple variables

sns.pairplot(data = df[['rating', 'discounted_price', 'actual_price', 'calculated_discount_filled']], diag_kind='kde')
````

#heatmap is generally created for correlation matrix (numeric values) - correlation means if one is increaing then the other value is also increasing or vice versa. Lower the spread more confidence you would have for more coorelation. In the following example B is greater then G in correlation.







image.png

````python
# heatmap
corr_matrix = df[['rating', 'discounted_price', 'actual_price', 'calculated_discount_filled']].corr()
plt.figure(figsize=(8, 6))

# Styling Tip 💡
# Format annotation to 2 decimal places
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='Blues')

plt.title("Correlation Heatmap", fontsize=16)
plt.show()
````
